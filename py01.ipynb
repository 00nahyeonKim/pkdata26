```python
# 응용
re = "{0:^18}".format("영수증")
print("=" * 20 + "\n" + re + "\n" + "=" * 20)
```

```python
americano_count = int(input("아아 개수 입력: "))
latte_count = int(input("라떼 개수 입력: " ))
print('')

americano_price = 2500
latte_price = 3800

total_americano = americano_count * americano_price
total_latte = latte_count * latte_price
total_price = total_americano + total_latte

print("=" * 20)
print("{:^18}".format("영수증"))
print("=" * 20)

print(f"{'아아':>8} {americano_count}잔")
print(f"{'라떼':>8} {latte_count}잔")

print("." * 20)

print(f"{'아아':>8} {total_americano}원")
print(f"{'라떼':>8} {total_latte}원")

print("-" * 20)

print(f">>>> 총가격 : {total_price} 원")
print("=" * 20)
```

## f 문자열 포매팅

```python
name = '홍길동'
age = 30

print(f"나의 이름은 {name}입니다. 나이는 {age} 입니다.")
```

```python
d = {'name': '홍길동', 'age': 30}

print(f"나의 이름은 {d['name'] }입니다. 나이는 {d['age']} 입니다.")
```

```python
d = {'name': '홍길동', 'age': 30}
# f포매팅에서 주의할 점
# 딕셔너리의 값을 표현할 때 괄호로 묶어 표기하고 공백을 주의하라.

print(f"나의 이름은 {(d['name']):^10}입니다. 나이는 {(d['age']):^10} 입니다.")
```

```python
# f문자열을 사용하여 세자리 콤마 찍기
btc = 76216
ssum = btc * 1450
f"{ssum:,}"
```
