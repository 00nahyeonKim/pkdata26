```python
import requests as req

f = open("즐겨찾기 종목.txt", "a")

url = "https://finance.naver.com/sise/sise_market_sum.naver"
ori = req.get(url).text

code = ['000270', '005930', '005380', '373220']

def finance(code = '005930'):
    start_title = ori.find(f'<a href="/item/main.naver?code={code}" class="tltle">')
    ori1 = ori[start_title : start_title + 70].replace(f'<a href="/item/main.naver?code={code}" class="tltle">', '').replace('</a></td>', '').replace('\n', '').strip()
    ori2 = ori[start_title + 70: start_title + 105].strip().replace('<td class="number">', '').replace('</td>', '').replace('</t', '')
    
    
    print("종목:", ori1)
    print("가격:", ori2)

    result = "종목:" + ori1 + "\n" + "가격:" + ori2 + "\n"
    f.write(result)
    
for x in code:
    finance(x)
f.close()
```

```python
# 파일 읽어오기. readline 함수 이용하기
import requests as req

url = "https://api4.binance.com/api/v3/ticker/price?symbol=BTCUSDT"
response_data = req.get(url).json()

price = float(response_data['price'])

won = str(int(price * 1450))
print(((won[-12:-8] + '억') if won[-12:-8] else ""), won[-8:-4]+ '만' ,won[-4:] + '원')
```

```python
import random as rand

f = open("/home/knh/data/lotto.txt", "w")

arr_lotto = []

for i in range(10):
    lotto = []
    
    while len(lotto) < 6:
        num = rand.randint(1, 100)
        if num not in lotto:
            lotto.append(num)
            
    arr_lotto.append(sorted(lotto))

for res in arr_lotto:
    f.write(str(res)+"\n")

f.close()
```

```python
import time as t

f = open("./즐겨찾기 종목.txt", "r")

while True:
    line = f.readline().strip()
    if not line: break
    print(line, end = "\r")
    t.sleep(1)
f.close()
```
