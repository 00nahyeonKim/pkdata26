```python
# 3번 조작
import random as rand

box = []
n = 0
cc = 10000
my = int(input("조작하고 싶은 숫자를 입력하세요."))
choice = int(input("조작할 확률 %를 입력하세요.") or 10) # 퍼센트 조절

while n < cc:
    num = rand.randint(1, 6)
    lucky = rand.randint(1, 100)
    if choice >= lucky:
        box.append(my)
    else:
        box.append(num)
    n += 1
    
if cc == len(box):
    print(f"{cc}회 정상 진행 되었습니다.")
    for i in range(1, 7):
        print(f"주사위 눈금 {i}이 {box.count(i)}번 {(box.count(i) / cc * 100):0.2f}% 출현하였습니다.")
else:
    print(f"검증 결과 이상이 있습니다. {cc}회 목표로 동작하였으나, {len(box)}으로 측정되었습니다.")
```

```python
# while 반복해서 1000 이상으로 돌려서 가장 많이 출현한 숫자를 뽑으세요

import random as rand

lst = []
lst2 = []

while True:
    num = rand.randint(1, 1001)
    lst.append(num)
    if len(lst) == 1000:
        break

for i in range(1, 1001):
    lst2.append(lst.count(i))

max_count = max(lst2)

winners = []

for i in range (len(lst2)):
    if lst2[i] == max_count:
        winners.append(i + 1)

print(f"가장 많이 나온 횟수: {max_count}")
print(f"해당 숫자들: {winners}")
```

```python
# 3번 조작
import random as rand
import matplotlib.pyplot as plt

box = []
n = 0
cc = 10000
values = []
my = int(input("조작하고 싶은 숫자를 입력하세요."))
choice = int(input("조작할 확률 %를 입력하세요.") or 10) # 퍼센트 조절

while n < cc:
    num = rand.randint(1, 6)
    lucky = rand.randint(1, 100)
    if choice >= lucky:
        box.append(my)
    else:
        box.append(num)
    n += 1
    
if cc == len(box):
    print(f"{cc}회 정상 진행 되었습니다.")
    for i in range(1, 7):
        print(f"주사위 눈금 {i}이 {box.count(i)}번 {(box.count(i) / cc * 100):0.2f}% 출현하였습니다.")
        values.append(box.count(i))
else:
    print(f"검증 결과 이상이 있습니다. {cc}회 목표로 동작하였으나, {len(box)}으로 측정되었습니다.")

color = ["skyblue", "orange"]

plt.bar(range(1, 7), values, color=color)
plt.show()
```
