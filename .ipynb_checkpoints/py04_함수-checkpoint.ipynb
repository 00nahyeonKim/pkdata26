```python
def add_many(*args):
    result = 0
    for i in args:
        result = result + i
    return result
```

```python
add_many(1, 2, 3, 4, 5)
```

```python
def add_mul(choice, *args):
    if choice == "add":
        result = 0
        for i in args:
            result = result + i
    elif choice == "mul":
        result = 1
        for i in args:
            result *= i
    return result
```

```python
result = add_mul('add', 1,2,3,4,5)
print(result)
result = add_mul('mul', 1,2,3,4,5)
print(result)
```

```python
# 위 예제에서 args를 하나만 넣고도 동작하도록 만들어보기

def add_mul(*args):
    if args[0] == "add":
        result = 0
        for i in args[1:]:
            result = result + i
    elif args[0] == "mul":
        result = 1
        for i in args[1:]:
            result *= i
    return result

result = add_mul('add', 1,2,3,4,5)
print(result)
result = add_mul('mul', 1,2,3,4,5)
print(result)
```

```python
result = add_mul('add', 1,2,3,4,5)
print(result)
result = add_mul('mul', 1,2,3,4,5)
print(result)
```

<fieldset><legend>점프 투 파이썬</legend><p style="margin-top: 0.2em;"><strong>리스트나 딕셔너리는 함수에서 변경 가능하다</strong></p>
<p>앞의 예제에서는 함수가 숫자나 문자열 같은 값을 받았다. 이런 값들은 함수 안에서 변경해도 함수 밖의 원래 값에는 영향을 주지 않는다. 하지만 리스트나 딕셔너리 같은 자료형은 다르다.</p>
<pre><code class="language-python hljs md-csc-wrap"><button class="md-csc-button btn btn-xs btn-default"><span class="copymsg">Copy</span></button><span class="hljs-meta">&gt;&gt;&gt; </span><span class="hljs-keyword">def</span> <span class="hljs-title function_">change_list</span>(<span class="hljs-params">my_list</span>):
<span class="hljs-meta">... </span>    my_list.append(<span class="hljs-number">4</span>)  <span class="hljs-comment"># 리스트에 값을 추가</span>
...
<span class="hljs-meta">&gt;&gt;&gt; </span>a = [<span class="hljs-number">1</span>, <span class="hljs-number">2</span>, <span class="hljs-number">3</span>]
<span class="hljs-meta">&gt;&gt;&gt; </span>change_list(a)
<span class="hljs-meta">&gt;&gt;&gt; </span><span class="hljs-built_in">print</span>(a)
[<span class="hljs-number">1</span>, <span class="hljs-number">2</span>, <span class="hljs-number">3</span>, <span class="hljs-number">4</span>]
</code></pre>
<p>위 예제를 보면 함수에서 리스트에 값을 추가했을 때 원래 리스트 a도 함께 변경되었다. 이는 리스트가 '변경 가능한(mutable)' 자료형이기 때문이다. 딕셔너리도 마찬가지로 함수 안에서 값을 변경하면 원래 딕셔너리도 함께 변경된다.</p>
<p>따라서 리스트나 딕셔너리를 함수에 전달할 때는 원본이 변경될 수 있다는 점을 기억하자.</p></fieldset>

```python
import random as rand

n, cc = 0, int(input("몇 회 주사위를 던질까요?")or 100)
box = []

while n < cc:
    num = rand.randint(1, 6)
    n += 1
    box.append(num)

if cc == len(box):
    print(f"{cc}회 정상 진행 되었습니다.")
    for i in range(1, 7):
        print(f"주사위 눈금 {i}이 {box.count(i)}번 {(box.count(i) / cc * 100):0.2f}% 출현하였습니다.")
else:
    print(f"검증 결과 이상이 있습니다. {cc}회 목표로 동작하였으나, {len(box)}으로 측정되었습니다.")
```

## lambda 예약어 *

```python
# 사용법
add = lambda a, b: a + b
result = add(11, 22)
print(result)
```

```python
# 한줄짜리 익명함수

print((lambda x: x**3)(2))
```

```python
# lambda 응용
students = [("철수", 90), ("영희", 85), ("민수", 95)]
ss = sorted(students, key = lambda x: x[1], reverse = True)
print(ss)
```

```python
import requests as req

url = "https://jsonplaceholder.typicode.com/todos"

result = req.get(url).json()

fin = sorted(result, key = lambda x : x['id'])

print("id를 기준으로 정렬")
print(fin)

completed_fin = [i for i in fin if i['completed'] == True]

print("\n\n\ncompleted가 True 된 내용만 보이게 전처리하라.\n\n\n")
print(completed_fin)
```

```python
import requests as req
import random as rand

url = "https://jsonplaceholder.typicode.com/todos"

result = req.get(url).json()

fix = [i | {"score" : rand.randint(40, 100)} for i in result]

# 리스트의 각 요소 끝에 score 이름으로 40~100까지의 임의 값으로 배정하여 fix라는 변수에 등록하라.
print(fix)

# score를 키로 내림차순으로 sort하라. (원본손상 x)

ss = sorted(fix, key = lambda x: x['score'])

print(ss)
```

```python
import requests as req

url = "https://finance.naver.com/sise/sise_market_sum.naver"

ori = req.get(url).text

start_title = ori.find('<a href="/item/main.naver?code=005930" class="tltle">삼성전자</a>')
ori1 = ori[start_title : start_title + 72].replace('<a href="/item/main.naver?code=005930" class="tltle">', '').replace('</a></td>', '').replace('\n', '').strip()
ori2 = ori[start_title + 72: start_title + 105].strip().replace('<td class="number">', '').replace('</td>', '')


start_title2 = ori.find('<a href="/item/main.naver?code=035420" class="tltle">')
ori11 = ori[start_title2 : start_title2 + 72].replace('<a href="/item/main.naver?code=035420" class="tltle">', '').replace('</a></td>', '').replace('\n', '').strip()
ori22 = ori[start_title2 + 72: start_title2 + 105].strip().replace('<td class="number">', '').replace('</td>', '')


print("종목:", ori1)
print("가격:", ori2)
print("종목:", ori11)
print("가격:", ori22)
```

```python
import requests as req

url = "https://finance.naver.com/sise/sise_market_sum.naver"
ori = req.get(url).text

code = ['000270', '005930', '005380', '373220']

def finance(code = '005930'):
    start_title = ori.find(f'<a href="/item/main.naver?code={code}" class="tltle">')
    ori1 = ori[start_title : start_title + 70].replace(f'<a href="/item/main.naver?code={code}" class="tltle">', '').replace('</a></td>', '').replace('\n', '').strip()
    ori2 = ori[start_title + 70: start_title + 105].strip().replace('<td class="number">', '').replace('</td>', '')
    
    
    print("종목:", ori1)
    print("가격:", ori2)
    
for x in code:
    finance(x)
```
