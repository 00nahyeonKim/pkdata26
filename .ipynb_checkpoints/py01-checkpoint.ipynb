```python
import requests as req

url = "https://api4.binance.com/api/v3/ticker/price?symbol=BTCUSDT"
response_data = req.get(url).json()

price = float(response_data['price'])

won = str(int(price * 1450))
print(won[:1] + '억' + won[1:5] + '만' + won[5:] + '원')
```

```python
# 문자열
print("%s" % '12.34 문자표현')
# 정수
print("%d" % 12.34)
# 부동소수점
print("%f" % 12.34)
# %%표기
print("한국에 산은 %d%% 정도 있습니다." % 70)
```

```python
# 소수점 표현하기

print("%.2f" % 3.141592)
print("%10.2fkkkk" % 3.141592)
print("%-10.2fkkkk" % 3.141592)
```

```python
s = 'five'
ss = "I eat {} apples".format(s)

print(sss, type(sss))
```

```python
s = 'five'
e = 7
ss = "I eat {0} apples, so I was sick for {1} days.".format(s, e)
print(ss, type(ss))
```

```python
sss = "I ate {0} apples. so i was sick for {day} days.".format(10, day = 3)
print(sss)
```

```python
# 왼쪽정렬
print("{0:<10}".format("hi"))
# 오른쪽 정렬
print("{0:>10}".format("hi"))
# 가운데 정렬
print("{0:^10}".format("hi"))
```

```python
# 응용
re = "{0:^18}".format("영수증")
print("=" * 20 + "\n" + re + "\n" + "=" * 20)
```
