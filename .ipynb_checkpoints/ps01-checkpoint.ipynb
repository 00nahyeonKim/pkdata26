```python
# 비트코인 API
import requests as req
url = "https://api.bithumb.com/v1/candles/minutes/1?market=KRW-BTC&count=1"
d_url = "https://api.bithumb.com/v1/candles/days?market=KRW-BTC&count=1"

ori = req.get(url).json()[0]
ori2 = req.get(d_url).json()[0]

# https://api.bithumb.com/v1/candles/minutes/1    물음표 앞까지 호스트에 요청하는 내용
# https://api.bithumb.com     호스트 주소
# market=KRW-BTC&count=1    "물음표 뒤에 내가 원하는걸 담아 보내겠다" 약속을 한 것
```

```python
print(ori, type(ori))
```

### 분봉

```python
# 시작가
openp = float(ori['opening_price'])
# 최고가
maxp = float(ori['high_price'])
# 최저가
minp = float(ori['low_price'])
# 변화폭 
bandp = maxp - minp

print("시작가: ", openp, "\n최고가: ", maxp, "\n최저가: ", minp, "\n변화폭: ", bandp)
# print("검토: ", openp + bandp)
# print("최고가: ", maxp)
result = "떡상중" if ((maxp - (openp + bandp)) > maxp) else "떡락중"
print("판단: ", result)
```

### 일봉

```python
# 시작가
openp2 = float(ori2['opening_price'])
# 최고가
maxp2 = float(ori2['high_price'])
# 최저가
minp2 = float(ori2['low_price'])
# 변화폭 
bandp2 = maxp2 - minp2

result2 = "떡상중" if ((maxp2 - (openp2 + bandp2)) > maxp2) else "떡락중"

print("판단: ", result2)
```

```python
# 판단하기

def check(me, comp_list):
    return me.lower() in [comp.lower() for comp in comp_list]
    
# 개인기술
Me = input("기술을 입력하세요: ")

# 모집공고
CompA = ['JAVA', "PYTHON", 'HTML', "Javascript", "C", "Oracle"]
CompB = ['HTML', "Python", "C", "MySQL"]
CompC = ['JAVA', 'HTML', "CSS", "Javascript", "C++", "Oracle"]

for i in [CompA, CompB, CompC]:
    print("합격" if check(Me, i) == True else "불합격")
```

```python
import requests as req
import webbrowser as wb
from datetime import datetime as dt

url = "https://api.open-meteo.com/v1/forecast?latitude=35.134032&longitude=129.1031735&current_weather=true"

raw = req.get(url).json()

raw_time = raw['current_weather']['time']
dt_obj = dt.strptime(raw_time, '%Y-%m-%dT%H:%M')
time = dt_obj.strftime('%Y년 %m월 %d일 %H시 %M분')

print("현재 시각: ", time)

temp = raw['current_weather']['temperature']

print("현재 온도: ", str(temp), '°C')

windspeed = raw['current_weather']['windspeed']

print('풍속: ', windspeed, 'm/s')

winddirection = raw['current_weather']['windspeed']

print('풍향: ', str(winddirection), '°')
```

```python
import requests as req

url = "https://api.fda.gov/drug/event.json?count=patient.reaction.reactionmeddrapt.exact"

raw = req.get(url).json()
data_dict = {item['term']: item['count'] for item in raw['results']}

nausea = data_dict['NAUSEA'] # 구역질
covid_19 = data_dict['COVID-19'] # 코로나-19
myocardial_infraction = data_dict['MYOCARDIAL INFARCTION'] # 심근경색
vomiting = data_dict['VOMITING'] # 구토

winner = "구역질" if nausea > vomiting else "구토"
diff = abs(nausea - vomiting) # 구역질과 구토의 차이

meta_data = raw['meta']['disclaimer']

print(f"구토증상의 횟수: {nausea:,}번")
print(f"COVID-19 관련 보고 건수: {covid_19:,}번")
print(f"MYOCARDIAL INFARCTION 건수: {myocardial_infraction:,}번")
print("코로나가 더 많음" if (covid_19 > myocardial_infraction) else "심근경색이 더 많음")
print(f"구역질과 구토를 비교했을 때, {winner}가 더 많으며 차이는 {diff:,}건임")
print("FDA가 들어있습니다" if 'FDA' in meta_data  else "들어있지 않습니다.")
```
